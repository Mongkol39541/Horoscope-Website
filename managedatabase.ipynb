```python
import pandas as pd
import numpy as np
import csv
from calculatePhone import phonenumber
```

```python
datafame = pd.read_csv('datacsv/phone.csv')
datafame
```

```python
loop = 0
for phone in datafame['Phone']:
    loop += 1
    if phone.isdigit() == False:
        datafame.drop([loop], axis=0, inplace=True)
datafame
```

```python
loop = 0
listloop = []
for num in datafame["Phone"]:
    listloop.append(loop)
    loop += 1
datafame["Num"] = listloop
max(datafame["PriceUSD"])
```

```python
datafame_Num = np.random.choice(datafame["Num"], 1000)
datafame_Num
```

```python
with open('datacsv/categoriesphone.csv', mode='w', encoding='utf-8') as csv_file:
    fieldnames = ['phone', 'price', 'double_num', 'category']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    for item in datafame_Num:
        price = int(datafame['PriceUSD'][item] * 30)
        total, mean_num, sum_percent, insum_percent, double_num, mean_duo_af, categorys = phonenumber(datafame["Phone"][item][:10])
        writer.writerow({'phone': str(datafame["Phone"][item][:10]), 'price': price, 'double_num': double_num, 'category': categorys})
```
