```python
import pandas as pd
import numpy as np
import csv
```

```python
datacate = pd.read_csv('datacsv/categoriesphone.csv')
datacate = pd.DataFrame(datacate)
datacate
```

```python
listcategory = []
newcategory = []
for txt in datacate["category"]:
    category = txt[:len(txt) - 1].split(",")
    newcategory.append(category)
    for che in category:
        listcategory.append(che)
set_category, num_category = np.unique(listcategory, return_counts=True)
print(set_category)
print(num_category)
```

```python
datacate["category"] = newcategory
datacate
```

```python
price_category = [0] * len(set_category)
data = datacate.shape
for num in range(data[0]):
    asix = 0
    for che in set_category:
        if che in datacate["category"][num]:
            price_category[asix] += int(datacate["price"][num])
        asix += 1
print(set_category)
print(price_category)
```

```python
listdouble = []
newdouble = []
for txt in datacate["double_num"]:
    double = txt[:len(txt) - 1].split(",")
    newdouble.append(double)
    for che in double:
        listdouble.append(che)
set_double, num_double = np.unique(listdouble, return_counts=True)
print(set_double)
print(num_double)
```

```python
datacate["double_num"] = newdouble
datacate
```

```python
price_double = [0] * len(set_double)
data = datacate.shape
for num in range(data[0]):
    asix = 0
    for che in set_double:
        if che in datacate["double_num"][num]:
            price_double[asix] += int(datacate["price"][num])
        asix += 1
print(set_double)
print(price_double)
```

```python
with open('datacsv/counts_category.csv', mode='w', encoding="utf-8") as csv_file:
    fieldnames = ['category', 'counts', 'price']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    for num in range(len(set_category)):
        writer.writerow({'category': set_category[num], 'counts': num_category[num], 'price': price_category[num]})
```

```python
with open('datacsv/counts_double.csv', mode='w') as csv_file:
    fieldnames = ['double', 'counts', 'price']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    for num in range(len(set_double)):
        writer.writerow({'double': set_double[num], 'counts': num_double[num], 'price': price_double[num]})
```
